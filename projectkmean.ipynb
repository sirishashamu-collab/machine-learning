K MEAN

```python
data.columns
```

```python
X=data[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
X.head()
```

```python
from sklearn.cluster import KMeans
```

```python
#it will only train the our algorithm
k_means=KMeans()
k_means.fit(X)
```

```python
#it will train and predict and create dependent variable
k_means=KMeans(n_clusters=5)
k_means.fit_predict(X)
```

ELBOW METHOD TO FIND OPTIMAL NUMBER OF CLUSTERS

```python
wcss=[]
for i in range(1,11):
    k_means=KMeans(n_clusters=i)
    k_means.fit(X)
    wcss.append(k_means.inertia_)
```

```python
wcss
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,11),wcss)
plt.title("ELBOW METHOD")
plt.xlabel("Number of Clusters")
plt.ylabel("wcss")
plt.show()
```

MODEL TRAINING

```python
X=data[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
k_means=KMeans(n_clusters=5,random_state=42)
y_means=k_means.fit_predict(X)
```

```python
y_means
```

```python
plt.scatter(X.iloc[y_means==0,0],X.iloc[y_means==0,1],s=50,c='green',label='Cluster1')
plt.scatter(X.iloc[y_means==1,0],X.iloc[y_means==1,1],s=50,c='red',label='Cluster2')
plt.scatter(X.iloc[y_means==2,0],X.iloc[y_means==2,1],s=50,c='black',label='Cluster3')
plt.scatter(X.iloc[y_means==3,0],X.iloc[y_means==3,1],s=50,c='yellow',label='Cluster4')
plt.scatter(X.iloc[y_means==4,0],X.iloc[y_means==4,1],s=50,c='purple',label='Cluster5')
plt.scatter(k_means.cluster_centers_[:,0],k_means.cluster_centers_[:,1],s=100,c='brown')
plt.title("Customer Segmentation")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X = data[['Annual Income (k$)','Spending Score (1-100)']].values

for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    labels = kmeans.fit_predict(X)   
    
    score = silhouette_score(X, labels)
    print(f"F={i}, Score={score}")
```
